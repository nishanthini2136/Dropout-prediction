```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display all columns
pd.set_option('display.max_columns', None)
```

```python
import os

dataset_path = "../Dataset"
```

```python
studentInfo = pd.read_csv(os.path.join(dataset_path, "studentInfo.csv"))

studentAssessment = pd.read_csv(os.path.join(dataset_path, "studentAssessment.csv"))

studentRegistration = pd.read_csv(os.path.join(dataset_path, "studentRegistration.csv"))

studentVle = pd.read_csv(os.path.join(dataset_path, "studentVle.csv"))

assessments = pd.read_csv(os.path.join(dataset_path, "assessments.csv"))

courses = pd.read_csv(os.path.join(dataset_path, "courses.csv"))

vle = pd.read_csv(os.path.join(dataset_path, "vle.csv"))
```

```python
datasets = {
    "studentInfo": studentInfo,
    "studentAssessment": studentAssessment,
    "studentRegistration": studentRegistration,
    "studentVle": studentVle,
    "assessments": assessments,
    "courses": courses,
    "vle": vle
}
```

```python
# Display Shape of Each Dataset
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print("Rows, Columns :", df.shape)
```

```python
# Display Column Names
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print(df.columns.tolist())
```

```python
# Display First Five Rows
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    display(df.head())
```

```python
# Display First Five Rows
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print(df.info())
```

```python
# Dataset Information
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print(df.isnull().sum())
```

```python
# Check Duplicate Rows
for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print("Duplicate Rows :", df.duplicated().sum())
```

```python
# Statistical Summary (Numeric Columns)
studentInfo.describe()
studentAssessment.describe()
studentVle.describe()
```

```python
# Check Target Variable
studentInfo["final_result"].value_counts()
```

```python
# Plot the Target Distribution
studentInfo["final_result"].value_counts().plot(kind="bar")

plt.title("Distribution of Final Result")
plt.xlabel("Final Result")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Save Exploration Summary
summary = []

for name, df in datasets.items():
    summary.append({
        "Dataset": name,
        "Rows": df.shape[0],
        "Columns": df.shape[1],
        "Missing Values": df.isnull().sum().sum(),
        "Duplicate Rows": df.duplicated().sum()
    })

summary_df = pd.DataFrame(summary)

summary_df
```
