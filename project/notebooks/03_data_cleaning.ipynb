```python
import pandas as pd
import numpy as np
import os
```

```python
dataset_path = "../Dataset"
```

```python
studentInfo = pd.read_csv(os.path.join(dataset_path, "studentInfo.csv"))

studentAssessment = pd.read_csv(os.path.join(dataset_path, "studentAssessment.csv"))

studentRegistration = pd.read_csv(os.path.join(dataset_path, "studentRegistration.csv"))

studentVle = pd.read_csv(os.path.join(dataset_path, "studentVle.csv"))

assessments = pd.read_csv(os.path.join(dataset_path, "assessments.csv"))

courses = pd.read_csv(os.path.join(dataset_path, "courses.csv"))

vle = pd.read_csv(os.path.join(dataset_path, "vle.csv"))
```

```python
datasets = {
    "studentInfo": studentInfo,
    "studentAssessment": studentAssessment,
    "studentRegistration": studentRegistration,
    "studentVle": studentVle,
    "assessments": assessments,
    "courses": courses,
    "vle": vle
}
```

```python
for name, df in datasets.items():
    print(name, df.shape)
```

```python
for name, df in datasets.items():
    print("\n", name)
    print(df.isnull().sum())
```

```python
for name, df in datasets.items():
    before = df.shape[0]

    df.drop_duplicates(inplace=True)

    after = df.shape[0]

    print(name, "Removed", before-after, "duplicate rows")
```

```python
studentRegistration["date_unregistration"] = studentRegistration["date_unregistration"].fillna(-1)
studentAssessment["score"] = studentAssessment["score"].fillna(
    studentAssessment["score"].median()
)
studentInfo["imd_band"] = studentInfo["imd_band"].fillna("Unknown")
```

```python
for name, df in datasets.items():
    print("\n", name)
    print(df.isnull().sum())
```

```python
studentInfo["dropout"] = studentInfo["final_result"].apply(
    lambda x: 1 if x=="Withdraw" else 0
)
```

```python
for name, df in datasets.items():
    print("\n", name)
    print(df.isnull().sum())
```

```python
studentInfo["dropout"] = studentInfo["final_result"].isin(
    ["Withdrawn"]
).astype(int)
```

```python
processed_path="../Dataset/processed"

os.makedirs(processed_path,exist_ok=True)
```

```python
studentInfo.to_csv(
    os.path.join(processed_path,"studentInfo_clean.csv"),
    index=False
)

studentAssessment.to_csv(
    os.path.join(processed_path,"studentAssessment_clean.csv"),
    index=False
)

studentRegistration.to_csv(
    os.path.join(processed_path,"studentRegistration_clean.csv"),
    index=False
)

studentVle.to_csv(
    os.path.join(processed_path,"studentVle_clean.csv"),
    index=False
)

assessments.to_csv(
    os.path.join(processed_path,"assessments_clean.csv"),
    index=False
)

courses.to_csv(
    os.path.join(processed_path,"courses_clean.csv"),
    index=False
)

vle.to_csv(
    os.path.join(processed_path,"vle_clean.csv"),
    index=False
)

print("All cleaned datasets saved successfully")
```
