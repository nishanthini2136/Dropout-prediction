```python
import pandas as pd
import glob
import os
```

```python
print(os.getcwd())
```

```python
dataset_path = "../Dataset"
os.listdir(dataset_path)
```

```python
vle_files = sorted(glob.glob(os.path.join(dataset_path, "studentVle_*.csv")))

print("Files Found:")
for file in vle_files:
    print(file)
```

```python
dataframes = []

for file in vle_files:
    print(f"Reading {file}")
    df = pd.read_csv(file)
    dataframes.append(df)

studentVle = pd.concat(dataframes, ignore_index=True)
```

```python
output_file = os.path.join(dataset_path, "studentVle.csv")

studentVle.to_csv(output_file, index=False)

print("Merged studentVle.csv saved successfully!")
```
